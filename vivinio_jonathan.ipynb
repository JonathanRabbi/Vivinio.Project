#### Question 1: Which 10 wines would you like to increase sales?

```python
# Which wines are the most expensive with highest rankings and from which year are they brewed?

query1 = """
SELECT DISTINCT vintage_id, name, year, ratings_average, ROUND (ratings_count,0),price_euros
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
ORDER BY rank ASC, ratings_average DESC, price_euros DESC
LIMIT 10;

    """
cursor.execute(query1)


data_1 = cursor.fetchall()
cols = ['vintage_id','name_wine', 'vintage', 'rating_average', 'rating_count', 'price_vintage']
df_1 = pd.DataFrame(data_1, columns=cols)
df_1
```

##### Relationships

```python
# Relationship between rank and ratings_average:

query2= """
SELECT DISTINCT name,rank,ratings_average
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
ORDER BY rank ASC, ratings_average DESC
LIMIT 70;

    """
cursor.execute(query2)
cursor.fetchall()
```

```python
# Relationship between rating_count and ratings_average:

query3= """
SELECT DISTINCT name,ratings_count,ratings_average
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
ORDER BY  ratings_count DESC
LIMIT 10;

    """
cursor.execute(query3)
cursor.fetchall()
```

#### Question 2: Which country should we prioritise and why?

```python
# We should prioritise Italy, as the most expensive wines vis-as-vis the ratings are the most premium.


query4 = """
SELECT 
    DISTINCT vintage_id,
    vintages.name, 
    year, 
    vintages.ratings_average, 
    ROUND (vintages.ratings_count,0),
    price_euros, 
    countries.name
    
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
    JOIN wines ON vintages.wine_id = wines.id
    JOIN regions ON wines.region_id = regions.id
    JOIN countries ON regions.country_code = countries.code

ORDER BY rank ASC, vintages.ratings_average DESC, price_euros DESC
LIMIT 10;

    """
cursor.execute(query4)

data_2 = cursor.fetchall()
cols = ['vintage_id','name_wine', 'vintage', 'rating_average', 'rating_count', 'price_vintage','countries']
df_2 = pd.DataFrame(data_2, columns=cols)
df_2
```
