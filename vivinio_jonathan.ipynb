#### Question 1: Which 10 wines would you like to increase sales?

```python
# Which wines are the most expensive with highest rankings and from which year are they brewed?

query1 = """
SELECT DISTINCT vintages.name, year, ratings_average, ROUND (ratings_count,0) AS count_rating, price_euros
FROM 
    vintages
    
WHERE count_rating > 500
ORDER BY  price_euros DESC, ratings_average DESC, ratings_count DESC


LIMIT 10;

    """
cursor.execute(query1)
data = cursor.fetchall()
cols = ['vintage_wine', 'vintage', 'rating_average', 'rating_count', 'price_vintage']
df_1 = pd.DataFrame(data, columns=cols)
df_1
```

```python
query2 = """
SELECT DISTINCT vintages.name, year, vintages.ratings_average, ROUND (vintages.ratings_count,0) AS count_rating, vintages.price_euros,countries.name
FROM 
    vintages
    JOIN wines ON vintages.wine_id = wines.id
    JOIN regions ON wines.region_id = regions.id
    JOIN countries ON regions.country_code = countries.code
    
WHERE count_rating > 500
ORDER BY  vintages.ratings_count DESC,vintages.ratings_average DESC


LIMIT 10;

    """
cursor.execute(query2)
data = cursor.fetchall()
cols = ['vintage_wine', 'vintage', 'rating_average', 'ratings_count', 'price_vintage','countries']
df_2 = pd.DataFrame(data, columns=cols)
df_2
```

```python
merged_df = pd.concat([df_1, df_2], ignore_index=True)
merged_df

```

#### Question 2: Which country should we prioritise and why?

```python
# We should prioritise Italy, as the most expensive wines vis-as-vis the ratings are the most premium.


query4 = """
SELECT 
    DISTINCT vintage_id,
    vintages.name, 
    year, 
    vintages.ratings_average, 
    ROUND (vintages.ratings_count,0),
    price_euros, 
    countries.name
    
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
    JOIN wines ON vintages.wine_id = wines.id
    JOIN regions ON wines.region_id = regions.id
    JOIN countries ON regions.country_code = countries.code

ORDER BY rank ASC, vintages.ratings_average DESC, price_euros DESC
LIMIT 10;

    """
cursor.execute(query4)

data = cursor.fetchall()
cols = ['vintage_id','name_wine', 'vintage', 'rating_average', 'rating_count', 'price_vintage','countries']
df_2 = pd.DataFrame(data, columns=cols)
df_2
```

#### Question 3: mixed tastes

```python
query5 = """
SELECT 
    DISTINCT vintages.name,
    countries.name,
    keywords.name,
    COUNT(keywords.name) AS number_mixed
  
    
FROM 
    vintage_toplists_rankings
    JOIN vintages ON vintage_toplists_rankings.vintage_id = vintages.id
    JOIN wines ON vintages.wine_id = wines.id
    JOIN regions ON wines.region_id = regions.id
    JOIN countries ON regions.country_code = countries.code
    JOIN keywords_wine ON wines.id = keywords_wine.wine_id
    JOIN keywords ON keywords_wine.keyword_id = keywords.id


WHERE 
    keywords.name ='coffee'
   AND  keywords.name ='toast'
   AND keywords.name='green apple'
   AND keywords.name='cream'
   AND keywords.name='citrus'

GROUP BY keywords.name;


    """
cursor.execute(query5)
data = cursor.fetchall()
cols = ['vintage_name', 'countries','mixed_taste','number_mixed_taste']
df_3 = pd.DataFrame(data, columns=cols)
df_3
```

```python
query7 = """
SELECT 
    countries.name, 
    AVG( vintages.ratings_average) AS average_rating_vintage,
    AVG(wines.ratings_average) AS average_rating_wine
    
FROM 
    vintages
    JOIN wines ON vintages.wine_id = wines.id
    JOIN regions ON wines.region_id = regions.id
    JOIN countries ON regions.country_code = countries.code

GROUP BY countries.name
ORDER BY average_rating_vintage DESC, average_rating_wine DESC;


    """
cursor.execute(query7)

data = cursor.fetchall()
cols = ['countries', 'rating_average_vintage','rating_average_wine']
df_2 = pd.DataFrame(data, columns=cols)
df_2
```
